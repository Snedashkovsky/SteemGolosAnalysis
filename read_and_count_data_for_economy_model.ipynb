## Load comments from MongoDB

```python
comment_df = pd.DataFrame(list(collection_comment.find({'_id' : {'$exists' : True}, 
                                                      'author' : {'$exists' : True},
                                                      'last_payout' : {'$exists' : True},
                                                      'reward_weight' : {'$exists' : True},
                                                      'net_rshares' : {'$exists' : True},
                                                      'net_votes' : {'$exists' : True},
                                                      'total_payout_value' : {'$exists' : True},
                                                      'root_comment' : {'$exists' : True},
                                                      'last_payout' : {'$exists' : True},
                                                      'depth' :{'$exists' : True},
                                                      'active_votes': {'$exists' : True}},                                              

                                                     {'_id': 1, 
                                                      'author': 1, 
                                                      'last_payout': 1, 
                                                      'reward_weight' : 1,
                                                      'net_rshares' : 1,
                                                      'net_votes' : 1,
                                                      'total_payout_value' : 1,
                                                      'curator_payout_value': 1,
                                                      'pending_payout_value': 1,
                                                      'max_accepted_payout':1,
                                                      'root_comment' : 1,
                                                      'last_payout' : 1,
                                                      'depth' : 1,
                                                      'active_votes':1})))
# comment_df.head()
```

```python
active_votes = comment_df['active_votes'][0]
def rshares_calculate(x):
    rshares = 0
    for items in x:
        rshares+=items['percent']*items['rshares']/PERCENT

    return rshares    
```

```python
comment_df.loc[:,'rshares'] = comment_df['active_votes'].map(rshares_calculate)
```

```python
comment_df.loc[:,'rshares'] = comment_df['rshares'] + comment_df['net_rshares']
```

```python
comment_df.loc[:,'payout_value'] = comment_df['total_payout_value']+ \
                                   comment_df['curator_payout_value']+ \
                                   comment_df['pending_payout_value']
```
